```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../trained_models/hls_quant/results.csv", index_col=0)
df
```

```python
fig, ax = plt.subplots(figsize=(5, 4))

ax.plot(df['fl'], df['test_acc'], label='Test Accuracy', marker='o')
ax.plot(df['fl'], df['train_acc'], label='Train Accuracy', marker='x')
ax.set_xlabel('Fraction Bit Length')
ax.set_ylabel('Accuracy')
ax.set_xticks(df['fl'])
ax.set_title('Accuracy vs Fraction Bit Length\n HLS High Level')
ax.legend()
fig.tight_layout()
fig.savefig('hls_quant_accuracy.png')
```
